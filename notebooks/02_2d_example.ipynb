# 2D Taylor Açılımı — e^(x₁²+x₂²)

Bu notebook, AOT paketini kullanarak iki değişkenli fonksiyonların 2. mertebe Taylor açılımını gösterir.

```python
import sympy as sp
import numpy as np
from aot import TaylorExpansion
```

## 1. Fonksiyon Tanımı

```python
x1, x2 = sp.symbols('x1 x2')
f = sp.exp(x1**2 + x2**2)

T = TaylorExpansion(f=f, variables=[x1, x2], point=[0, 0], order=2)
print("Fonksiyon: e^(x1² + x2²)")
print("Açılım noktası: (0, 0)")
```

## 2. Sembolik Hesap

```python
print("Taylor açılımı:")
T  # LaTeX render
```

```python
print("Gradient (∇f):\n", T.gradient)
print("\nHessian (H):\n", T.hessian)
```

## 3. Sayısal Karşılaştırma

```python
import pandas as pd

f_num = T.to_numeric()
t_num = T.taylor_to_numeric()

test_points = [(0, 0), (0.1, 0.1), (0.3, 0.2), (0.5, 0.5), (1.0, 0.0)]
rows = []
for x1v, x2v in test_points:
    fv = float(f_num(x1v, x2v))
    tv = float(t_num(x1v, x2v))
    rows.append({"x1": x1v, "x2": x2v, "f(x)": fv, "T(x)": tv, "|f-T|": abs(fv - tv)})

pd.DataFrame(rows).round(6)
```

## 4. Görselleştirme (3D Yüzey)

```python
# Plotly varsa interaktif, yoksa matplotlib kullanılır
fig = T.plot(xlim=(-1.5, 1.5), ylim=(-1.5, 1.5))
fig
```

## 5. Hata Analizi

```python
import matplotlib.pyplot as plt

r_vals = np.linspace(0, 1.5, 200)  # orijinden uzaklık
# 45 derece yönünde kesit: x1=x2=r/√2
x1_line = r_vals / np.sqrt(2)
x2_line = r_vals / np.sqrt(2)

error = np.abs(f_num(x1_line, x2_line) - t_num(x1_line, x2_line))

fig, ax = plt.subplots(figsize=(8, 4))
ax.semilogy(r_vals, error, color='purple')
ax.set_xlabel('r (orijinden uzaklık, 45° yönünde)')
ax.set_ylabel('|f(x) - T(x)|')
ax.set_title('Hata Analizi: e^(x1²+x2²) Taylor Açılımı')
ax.grid(True, alpha=0.3)
fig
```
