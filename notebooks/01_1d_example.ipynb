# 1D Taylor Açılımı — sin(x), e^x ve from_latex

Bu notebook, AOT paketini kullanarak tek değişkenli fonksiyonların yüksek mertebe Taylor açılımını gösterir.

**Yenilikler (v0.2):** `order` 1–1000 arası serbest, `from_latex` ile LaTeX string girişi, `latex(max_terms)` ile kısaltılmış çıktı.

```python
import sympy as sp
import numpy as np
from aot import TaylorExpansion
```

## 1. Fonksiyon Tanımı

```python
x = sp.Symbol('x')
f = sp.exp(x)

# order=5 ile daha iyi yaklaşım
T = TaylorExpansion(f=f, variables=[x], point=[0], order=5)
print("Fonksiyon:", f)
print("Açılım noktası: x=0, order=5")
```

## 2. Sembolik Hesap

```python
print("Taylor açılımı (ilk 5 terim):")
print(T.latex(max_terms=5))
print()
T  # Jupyter'da LaTeX olarak render edilir
```

```python
# 1D'de gradient ve hessian None döner — sp.series() yolu kullanılır
print("1D için gradient/hessian: gradient =", T.gradient, "/ hessian =", T.hessian)
print("Sembolik ifade:", T.symbolic)
```

## 2b. from_latex ile Giriş (yeni v0.2)

```python
# LaTeX string ile doğrudan giriş
T_latex = TaylorExpansion.from_latex(r"\sin(x)", point=[0], order=7)
print("from_latex ile sin(x), order=7:")
print(T_latex.latex(max_terms=4))
T_latex
```

## 3. Sayısal Karşılaştırma

```python
import pandas as pd

f_num = T.to_numeric()
t_num = T.taylor_to_numeric()

test_points = [-1.0, -0.5, 0.0, 0.5, 1.0]
rows = []
for xv in test_points:
    fv = float(f_num(xv))
    tv = float(t_num(xv))
    rows.append({"x": xv, "f(x)": fv, "T(x)": tv, "|f-T|": abs(fv - tv)})

pd.DataFrame(rows).round(6)
```

## 5. Hata Analizi

```python
import matplotlib.pyplot as plt
import numpy as np

x_vals = np.linspace(-2, 2, 400)
error = np.abs(f_num(x_vals) - t_num(x_vals))

fig, ax = plt.subplots(figsize=(8, 4))
ax.semilogy(x_vals, error, color='purple')
ax.set_xlabel('x')
ax.set_ylabel('|f(x) - T(x)|')
ax.set_title('Hata Analizi: e^x Taylor Açılımı')
ax.grid(True, alpha=0.3)
fig
```
