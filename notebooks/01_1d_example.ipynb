# 1D Taylor Açılımı — sin(x) ve e^x

Bu notebook, AOT paketini kullanarak tek değişkenli fonksiyonların 2. mertebe Taylor açılımını gösterir.

```python
import sympy as sp
import numpy as np
from aot import TaylorExpansion
```

## 1. Fonksiyon Tanımı

```python
x = sp.Symbol('x')
f = sp.exp(x)
T = TaylorExpansion(f=f, variables=[x], point=[0], order=2)
print("Fonksiyon:", f)
print("Açılım noktası: x=0")
```

## 2. Sembolik Hesap

```python
print("Taylor açılımı:")
T  # Jupyter'da LaTeX olarak render edilir
```

```python
print("Gradient:", T.gradient)
print("Hessian :", T.hessian)
```

## 3. Sayısal Karşılaştırma

```python
import pandas as pd

f_num = T.to_numeric()
t_num = T.taylor_to_numeric()

test_points = [-1.0, -0.5, 0.0, 0.5, 1.0]
rows = []
for xv in test_points:
    fv = float(f_num(xv))
    tv = float(t_num(xv))
    rows.append({"x": xv, "f(x)": fv, "T(x)": tv, "|f-T|": abs(fv - tv)})

pd.DataFrame(rows).round(6)
```

## 4. Görselleştirme

```python
%matplotlib inline
fig = T.plot(xlim=(-2, 2))
fig
```

## 5. Hata Analizi

```python
import matplotlib.pyplot as plt
import numpy as np

x_vals = np.linspace(-2, 2, 400)
error = np.abs(f_num(x_vals) - t_num(x_vals))

fig, ax = plt.subplots(figsize=(8, 4))
ax.semilogy(x_vals, error, color='purple')
ax.set_xlabel('x')
ax.set_ylabel('|f(x) - T(x)|')
ax.set_title('Hata Analizi: e^x Taylor Açılımı')
ax.grid(True, alpha=0.3)
fig
```
